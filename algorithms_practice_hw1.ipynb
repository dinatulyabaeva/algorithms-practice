```python
#Развернуть часть массива
def rotate_array(nums, k):
    n = len(nums)
    k %= n  #если k>n

    def reverse(arr, left, right):
        while left < right:
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1

    reverse(nums, 0, n - 1)
    reverse(nums, 0, k - 1)
    reverse(nums, k, n - 1)
```

```python
#Слияние двух отсортированных массивов
def merge_sorted(arr1, arr2):
    i = j = 0
    result = []

    while i < len(arr1) and j < len(arr2):
        if arr1[i] <= arr2[j]:
            result.append(arr1[i])
            i += 1
        else:
            result.append(arr2[j])
            j += 1

    result.extend(arr1[i:])
    result.extend(arr2[j:])
    return result
```

```python
#Слияние двух отсортированных массивов (без дополнительных аллокаций)
def merge_in_place(nums1, m, nums2, n):
    i = m - 1
    j = n - 1
    k = m + n - 1

    while i >= 0 and j >= 0:
        if nums1[i] > nums2[j]:
            nums1[k] = nums1[i]
            i -= 1
        else:
            nums1[k] = nums2[j]
            j -= 1
        k -= 1

    while j >= 0:
        nums1[k] = nums2[j]
        j -= 1
        k -= 1
```

```python
#Сортировка массива из 0 и 1
def sort_zeros_ones(nums):
    zero_pos = 0

    for i in range(len(nums)):
        if nums[i] == 0:
            nums[i], nums[zero_pos] = nums[zero_pos], nums[i]
            zero_pos += 1
```

```python
#Задача флага Нидерландов
def dutch_flag(nums):
    low = 0
    mid = 0
    high = len(nums) - 1

    while mid <= high:
        if nums[mid] == 0:
            nums[low], nums[mid] = nums[mid], nums[low]
            low += 1
            mid += 1
        elif nums[mid] == 1:
            mid += 1
        else:
            nums[mid], nums[high] = nums[high], nums[mid]
            high -= 1
```

```python
#Передвинуть чётные числа вперёд
def move_even_forward(nums):
    insert_pos = 0

    for i in range(len(nums)):
        if nums[i] % 2 == 0:
            nums[insert_pos], nums[i] = nums[i], nums[insert_pos]
            insert_pos += 1
```

```python
#Нули в конец
def move_zeros(nums):
    insert_pos = 0

    for i in range(len(nums)):
        if nums[i] != 0:
            nums[insert_pos] = nums[i]
            insert_pos += 1

    while insert_pos < len(nums):
        nums[insert_pos] = 0
        insert_pos += 1
```

```python
def run_tests():
    print("запуск тестов...\n")

    nums = [1, 2, 3, 4, 5, 6, 7]
    rotate_array(nums, 3)
    assert nums == [5, 6, 7, 1, 2, 3, 4]
    print("rotate_array работает корректно")

    result = merge_sorted([1, 3, 5], [2, 4, 6])
    assert result == [1, 2, 3, 4, 5, 6]
    print("merge_sorted работает корректно")

    nums1 = [1, 2, 3, 0, 0, 0]
    merge_in_place(nums1, 3, [2, 5, 6], 3)
    assert nums1 == [1, 2, 2, 3, 5, 6]
    print("merge_in_place работает корректно")

    nums = [1, 0, 1, 0, 1, 0]
    sort_zeros_ones(nums)
    assert nums == [0, 0, 0, 1, 1, 1]
    print("sort_zeros_ones работает корректно")

    nums = [2, 0, 2, 1, 1, 0]
    dutch_flag(nums)
    assert nums == [0, 0, 1, 1, 2, 2]
    print("dutch_flag работает корректно")

    nums = [3, 2, 4, 1, 11, 8, 9]
    move_even_forward(nums)
    assert nums == [2, 4, 8, 1, 11, 3, 9]
    print("move_even_forward работает корректно")

    nums = [0, 1, 0, 3, 12]
    move_zeros(nums)
    assert nums == [1, 3, 12, 0, 0]
    print("move_zeros работает корректно")

run_tests()
```

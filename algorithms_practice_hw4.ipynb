```python
from collections import deque


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
#Восстановление бинарного дерева из массива
def build_tree_from_level_order(arr):
    if not arr or arr[0] is None:
        return None

    root = TreeNode(arr[0])
    queue = deque([root])
    i = 1

    while queue and i < len(arr):
        current = queue.popleft()

        if i < len(arr) and arr[i] is not None:
            current.left = TreeNode(arr[i])
            queue.append(current.left)
        i += 1

        if i < len(arr) and arr[i] is not None:
            current.right = TreeNode(arr[i])
            queue.append(current.right)
        i += 1

    return root
```

```python
#Симметричное бинарное дерево
def is_symmetric(root):
    def is_mirror(left, right):
        if left is None and right is None:
            return True
        if left is None or right is None:
            return False
        if left.val != right.val:
            return False

        return is_mirror(left.left, right.right) and is_mirror(left.right, right.left)

    if root is None:
        return True

    return is_mirror(root.left, root.right)
```

```python
#Поиск минимальной глубины бинарного дерева
def min_depth(root):
    if root is None:
        return 0

    queue = deque([(root, 1)])

    while queue:
        node, depth = queue.popleft()

        if node.left is None and node.right is None:
            return depth

        if node.left:
            queue.append((node.left, depth + 1))
        if node.right:
            queue.append((node.right, depth + 1))
```

```python
#Поиск произведения максимального и минимального элементов
def find_min_in_bst(root):
    current = root
    while current and current.left:
        current = current.left
    return current.val if current else None


def find_max_in_bst(root):
    current = root
    while current and current.right:
        current = current.right
    return current.val if current else None


def min_max_product_bst(arr):
    root = build_tree_from_level_order(arr)
    if root is None:
        return None

    min_val = find_min_in_bst(root)
    max_val = find_max_in_bst(root)

    return min_val * max_val
```

```python
#Сравнение двух бинарных деревьев
def is_same_tree(p, q):
    if p is None and q is None:
        return True
    if p is None or q is None:
        return False
    if p.val != q.val:
        return False

    return is_same_tree(p.left, q.left) and is_same_tree(p.right, q.right)
```

```python
def tree_to_level_order(root):
    if not root:
        return []

    result = []
    queue = deque([root])

    while queue:
        node = queue.popleft()

        if node is None:
            result.append(None)
        else:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)

    while result and result[-1] is None:
        result.pop()

    return result
```

```python
print("Тесты для build_tree_from_level_order")

arr = [1, 2, 3, 4, 5, None, 6]
root = build_tree_from_level_order(arr)
print(tree_to_level_order(root))  # [1, 2, 3, 4, 5, None, 6]

arr2 = []
root2 = build_tree_from_level_order(arr2)
print(tree_to_level_order(root2))  # []

arr3 = [1]
root3 = build_tree_from_level_order(arr3)
print(tree_to_level_order(root3))  # [1]
```

```python
print("\nТесты для is_symmetric")

root1 = build_tree_from_level_order([1, 2, 2, 3, 4, 4, 3])
print(is_symmetric(root1))  # True

root2 = build_tree_from_level_order([1, 2, 2, None, 3, None, 3])
print(is_symmetric(root2))  # False

root3 = build_tree_from_level_order([])
print(is_symmetric(root3))  # True
```

```python
print("\nТесты для min_depth")

root1 = build_tree_from_level_order([3, 9, 20, None, None, 15, 7])
print(min_depth(root1))  # 2

root2 = build_tree_from_level_order([1, 2, None, 3, None, 4, None])
print(min_depth(root2))  # 4

root3 = build_tree_from_level_order([])
print(min_depth(root3))  # 0

root4 = build_tree_from_level_order([1])
print(min_depth(root4))  # 1
```

```python
print("\nТесты для min_max_product_bst")

bst1 = [5, 3, 8, 2, 4, 7, 9]
print(min_max_product_bst(bst1))  # 2 * 9 = 18

bst2 = [10, 5, 15, 2, 7, None, 20]
print(min_max_product_bst(bst2))  # 2 * 20 = 40

bst3 = [6]
print(min_max_product_bst(bst3))  # 36
```

```python
print("\nТесты для is_same_tree")

tree1 = build_tree_from_level_order([1, 2, 3])
tree2 = build_tree_from_level_order([1, 2, 3])
print(is_same_tree(tree1, tree2))  # True

tree3 = build_tree_from_level_order([1, 2])
tree4 = build_tree_from_level_order([1, None, 2])
print(is_same_tree(tree3, tree4))  # False

tree5 = build_tree_from_level_order([])
tree6 = build_tree_from_level_order([])
print(is_same_tree(tree5, tree6))  # True
```

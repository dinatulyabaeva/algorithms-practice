```python
#Максимальная сумма подмассива
def max_sum_subarray_k(arr, k):
    n = len(arr)

    if k > n:
        return None

    prefix = [0] * (n + 1)

    for i in range(n):
        prefix[i + 1] = prefix[i] + arr[i]

    max_sum = float("-inf")

    for i in range(n - k + 1):
        current_sum = prefix[i + k] - prefix[i]
        max_sum = max(max_sum, current_sum)

    return max_sum


print(max_sum_subarray_k([1, 2, 3, 4, 5], 3))  # 12
print(max_sum_subarray_k([5, -2, 3, 1, 2], 2)) # 4
```

```python
#Subarray Sum Equals K
def subarray_sum_equals_k(nums, k):
    prefix_count = {0: 1}
    prefix_sum = 0
    result = 0

    for num in nums:
        prefix_sum += num

        need = prefix_sum - k

        if need in prefix_count:
            result += prefix_count[need]

        prefix_count[prefix_sum] = prefix_count.get(prefix_sum, 0) + 1

    return result


print(subarray_sum_equals_k([1, 1, 1], 2))      # 2
print(subarray_sum_equals_k([1, 2, 3], 3))      # 2
print(subarray_sum_equals_k([3, 4, 7, 2], 7))   # 2
```

```python
#Максимальная длина подмассива с равным количеством нулей и единиц
def max_len_equal_zero_one(nums):
    first_position = {0: -1}
    prefix_sum = 0
    max_len = 0

    for i, num in enumerate(nums):
        if num == 0:
            prefix_sum += -1
        else:
            prefix_sum += 1

        if prefix_sum in first_position:
            current_len = i - first_position[prefix_sum]
            max_len = max(max_len, current_len)
        else:
            first_position[prefix_sum] = i

    return max_len


print(max_len_equal_zero_one([0, 1]))              # 2
print(max_len_equal_zero_one([0, 1, 0]))           # 2
print(max_len_equal_zero_one([0, 0, 1, 1, 0, 1]))  # 6
```

```python
#Индекс поворота массива
def find_pivot_index(nums):
    left = 0
    right = len(nums) - 1

    while left < right:
        mid = (left + right) // 2

        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid

    return left


print(find_pivot_index([4, 5, 6, 7, 0, 1, 2]))  # 4
print(find_pivot_index([3, 4, 5, 1, 2]))        # 3
print(find_pivot_index([1, 2, 3, 4]))           # 0
```

```python
#Баланс скобок через префиксные суммы
def is_valid_parentheses(s):
    balance = 0

    for char in s:
        if char == "(":
            balance += 1
        elif char == ")":
            balance -= 1

        if balance < 0:
            return False

    return balance == 0


print(is_valid_parentheses("(()())"))  # True
print(is_valid_parentheses("())("))    # False
print(is_valid_parentheses("((())"))   # False
```

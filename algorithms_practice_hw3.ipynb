```python
#1. Найти корень числа
def integer_sqrt(n: int) -> int:
    if n < 0:
        raise ValueError("корня для отрицательного числа нет")

    if n == 0 or n == 1:
        return n

    left, right = 0, n
    answer = 0

    while left <= right:
        mid = (left + right) // 2
        square = mid * mid

        if square == n:
            return mid
        elif square < n:
            answer = mid
            left = mid + 1
        else:
            right = mid - 1

    return answer
```

```python
#Очень лёгкая задача (два ксерокса)
def min_copy_time(n: int, x: int, y: int) -> int:
    if n <= 0:
        return 0
    if n == 1:
        return min(x, y)

    first_copy = min(x, y)
    left = 0
    right = max(x, y) * (n - 1)

    while left <= right:
        mid = (left + right) // 2

        copies = mid // x + mid // y

        if copies >= n - 1:
            right = mid - 1
        else:
            left = mid + 1

    return first_copy + left
```

```python
#Накормить животных
def feed_animals(animals, portions):
    animals.sort()
    portions.sort()

    i = 0
    j = 0
    fed = 0

    while i < len(animals) and j < len(portions):
        if portions[j] >= animals[i]:
            fed += 1
            i += 1
            j += 1
        else:
            j += 1

    return fed
```

```python
#Найти разницу между двумя строками
def find_added_char_count(a: str, b: str) -> str:
    counts = {}

    for ch in b:
        counts[ch] = counts.get(ch, 0) + 1

    for ch in a:
        counts[ch] -= 1

    for ch, count in counts.items():
        if count > 0:
            return ch
```

```python
#Сумма двух элементов массива
def two_sum(nums, target):
    seen = {}

    for i, num in enumerate(nums):
        needed = target - num

        if needed in seen:
            return [seen[needed], i]

        seen[num] = i

    return []
```

```python
# 1. integer_sqrt
print("1. integer_sqrt")
print(integer_sqrt(0))    # 0
print(integer_sqrt(1))    # 1
print(integer_sqrt(8))    # 2
print(integer_sqrt(16))   # 4
print(integer_sqrt(27))   # 5

# 2. min_copy_time
print("\n2. min_copy_time")
print(min_copy_time(1, 1, 2))   # 1
print(min_copy_time(4, 1, 1))   # 3
print(min_copy_time(5, 1, 2))   # 4

# 3. feed_animals
print("\n3. feed_animals")
print(feed_animals([1, 2, 3], [1, 1]))         # 1
print(feed_animals([1, 2], [1, 2, 3]))         # 2
print(feed_animals([2, 3, 4], [1, 2, 3]))      # 2

# 4. find_added_char
print("\n4. find_added_char")
print(find_added_char_count("abcd", "abcde"))        # e
print(find_added_char_count("aabb", "ababc"))        # c
print(find_added_char_count("", "z"))                # z

# 5. two_sum
print("\n5. two_sum")
print(two_sum([2, 7, 11, 15], 9))              # [0, 1]
print(two_sum([3, 2, 4], 6))                   # [1, 2]
print(two_sum([1, 2, 3], 7))                   # []
```

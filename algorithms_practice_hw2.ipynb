```python
def create_linked_list(arr):
    if not arr:
        return None

    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next

    return head


def linked_list_to_list(head):
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
#Развернуть односвязный список
def reverse_list(head: ListNode) -> ListNode:
    prev = None
    current = head

    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node

    return prev
```

```python
head = create_linked_list([1, 2, 3, 4])
reversed_head = reverse_list(head)
print(linked_list_to_list(reversed_head))

head = create_linked_list([10])
print(linked_list_to_list(reverse_list(head)))

print(reverse_list(None))
```

```python
#Найти середину списка (метод двух указателей)
def find_middle(head: ListNode) -> ListNode:
    slow = head
    fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    return slow
```

```python
head = create_linked_list([1, 2, 3, 4, 5])
middle = find_middle(head)
print(middle.val)

head = create_linked_list([1, 2, 3, 4])
middle = find_middle(head)
print(middle.val)
```

```python
#Удалить элемент из списка
def delete_node(head: ListNode, val: int) -> ListNode:
    if head and head.val == val:
        return head.next

    current = head

    while current and current.next:
        if current.next.val == val:
            current.next = current.next.next
            break
        current = current.next

    return head
```

```python
head = create_linked_list([1, 2, 3, 4])
head = delete_node(head, 3)
print(linked_list_to_list(head))  # [1, 2, 4]

head = create_linked_list([5, 6, 7])
head = delete_node(head, 5)
print(linked_list_to_list(head))  # [6, 7]

head = create_linked_list([1, 2])
head = delete_node(head, 10)
print(linked_list_to_list(head))  # [1, 2]
```

```python
#Является ли строка a исходной для строки b

def is_original(a: str, b: str) -> bool:
    i = 0
    j = 0

    while i < len(a) and j < len(b):
        if a[i] == b[j]:
            i += 1
        j += 1

    return i == len(a)
```

```python
print(is_original("abc", "aXbYcZ"))
print(is_original("abc", "acb"))
print(is_original("", "abc"))
print(is_original("abc", "ab"))
```

```python
#Поиск пары в отсортированном массиве
def has_pair(arr, target):
    left = 0
    right = len(arr) - 1

    while left < right:
        current_sum = arr[left] + arr[right]

        if current_sum == target:
            return True
        elif current_sum < target:
            left += 1
        else:
            right -= 1

    return False
```

```python
print(has_pair([1, 2, 3, 4, 6], 6))
print(has_pair([1, 2, 3, 4], 10))
print(has_pair([-2, 0, 3, 5], 3))
```

```python
#Проверка палиндрома
def is_palindrome(s: str) -> bool:
    left = 0
    right = len(s) - 1

    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1

    return True
```

```python
print(is_palindrome("racecar"))
print(is_palindrome("abba"))
print(is_palindrome("hello"))
print(is_palindrome("a"))
```

```python
#Удаление дубликатов в отсортированном массиве
def remove_duplicates(arr):
    if not arr:
        return 0

    slow = 0

    for fast in range(1, len(arr)):
        if arr[fast] != arr[slow]:
            slow += 1
            arr[slow] = arr[fast]

    return slow + 1
```

```python
arr = [1, 1, 2, 2, 3, 4, 4]
k = remove_duplicates(arr)
print(arr[:k])  # [1, 2, 3, 4]

arr = [1, 1, 1]
k = remove_duplicates(arr)
print(arr[:k])

arr = []
k = remove_duplicates(arr)
print(k)
```

```python
#Слияние двух отсортированных списков
def merge_lists(list1: ListNode, list2: ListNode) -> ListNode:
    dummy = ListNode()
    current = dummy

    while list1 and list2:
        if list1.val < list2.val:
            current.next = list1
            list1 = list1.next
        else:
            current.next = list2
            list2 = list2.next

        current = current.next

    if list1:
        current.next = list1
    if list2:
        current.next = list2

    return dummy.next
```

```python
list1 = create_linked_list([3, 6, 8])
list2 = create_linked_list([4, 7, 9, 11])

merged = merge_lists(list1, list2)
print(linked_list_to_list(merged))

list1 = None
list2 = create_linked_list([1, 2])
merged = merge_lists(list1, list2)
print(linked_list_to_list(merged))
```
